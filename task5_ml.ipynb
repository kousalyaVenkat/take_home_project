Creating new target variable adopted in users data frame using data from engagement data

```python
time_df = pd.DataFrame(df_engagement.set_index('time_stamp')
                       .groupby('user_id')['visited'].resample('D').count().reset_index())
time_df.drop_duplicates(
    subset=['user_id', 'time_stamp'], keep='first', inplace=True)
time_df = pd.DataFrame(df_engagement.set_index('time_stamp')
                       .groupby('user_id')['visited'].resample('7D').count().reset_index())
adopted_user_df = pd.DataFrame(time_df[time_df['visited'] >= 3]['user_id']
                               .unique().tolist(), columns=['object_id'])
adopted_user_df['adopted'] = 1
df_users = df_users.join(
    adopted_user_df.set_index('object_id'))
df_users['adopted'].fillna(0, inplace=True)
df_users['adopted'] = df_users['adopted'].astype(int)
df_users
```

```python
df = df_users[['creation_source', 'opted_in_to_mailing_list',
              'enabled_for_marketing_drip', 'org_id', 'invited_by_user_id', 'adopted']]
```

```python
sns.heatmap(df.corr())
```

```python
df = df_users[[ 'creation_source',  'org_id', 'invited_by_user_id', 'adopted']]
df['opted_marketing'] = df_users['opted_in_to_mailing_list'] + \
    df_users['enabled_for_marketing_drip']
df
```

```python
print(df.columns)
features = ['creation_source', 'org_id',
            'invited_by_user_id', 'opted_marketing']
for feature in features:
    print(f'{feature}: {len(df[feature].unique())}')
```

```python
import matplotlib.pyplot as plt
```

```python
features = ['creation_source', 'opted_marketing']


def createPlot(data, feature, axis, title):
    sns.barplot(data=data, ax=axis, x=feature, y='adopted', color='steelblue')
    axis.set_title(title)


rows, cols = 1, 2
fig, axes = plt.subplots(rows, cols, figsize=(18, 10))
fig.subplots_adjust(wspace=0.3, hspace=0.3)

for i, feature in enumerate(features):
    data = df.groupby(feature)['adopted'].sum().reset_index()
    createPlot(data, feature, axes[i], f'{feature} vs adopted')
```

opted_marketing is in correct ordinal/label encoding

```python
features = ['org_id', 'invited_by_user_id']


def createPlot(data, axis, title):
    sns.kdeplot(data=data, ax=axis, fill='steelblue')
    axis.set_title(title)


rows, cols = 1, 2
fig, axes = plt.subplots(rows, cols, figsize=(18, 10))
for i, feature in enumerate(features):
    data = df.groupby(feature)['adopted'].sum().reset_index()
    createPlot(data['adopted'], axes[i], f'{feature} vs adopted')
```

From above analysis we can determine that org_id and invite_by_user_id has very high unique variables . So we will drop org_id and convert invite_by_user_id to a binary encoding

We can also create new feature for number of people referred called qty_referred using info in invite_by_user_id column

```python
qty_referred = df_users.groupby('invited_by_user_id')['invited_by_user_id']\
    .count().rename('qty_referred')
df = df.join(qty_referred)
df['qty_referred'].fillna(0, inplace=True)
df['qty_referred'] = df['qty_referred'].astype(np.int16) 
df
```

```python
df.rename(columns={'invited_by_user_id': 'invited_by_user'}, inplace=True)
df['invited_by_user'] = pd.Series(
    df_users['invited_by_user_id'] >= 0, dtype=np.int0)
df
```

```python
df = df.drop('org_id', axis=1)
df
```

```python
df_temp = df_users.groupby('creation_source')['adopted']\
    .sum().sort_values().reset_index()['creation_source']
dict_ = df_temp.to_dict()
replace = dict( zip( dict_.values(),dict_.keys() ) )
df['creation_source'] = df_users['creation_source'].replace(replace)
df
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('adopted', axis=1)
y = df['adopted'] 

X_train, X_test, y_train, y_test = train_test_split(\
     X, y, test_size=0.3, random_state=0)
```

```python
from sklearn import svm

clf = svm.SVC() 

clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = clf.predict(X_test)

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
(tn, fp, fn, tp)
```

But the confusion matrix shows that our model always predicted the result as false and because there is low number of positive cases it has received a high score which is misleading and we need to tune our features and model selection for better prediction
